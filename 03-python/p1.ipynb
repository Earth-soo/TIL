# 문제 1. 점수 평균과 합격 판별
학생 점수 리스트가 있다.

scores = [80, 75, 90, 60, 100]

평균을 구하시오. 
평균이 80 이상이면 "합격" 아니면 "불합격" 출력

```python
scores = [80, 75, 90, 60, 100]
sum = 0

for s in scores:
    sum += s

avg = sum / len(scores)
print(avg)

if avg >= 80:
    print('합격')
else:
    print('불합격')
```

# 문제 2. 최대값 찾기
숫자 리스트가 있다.

numbers = [3, 7, 2, 9, 5]

가장 큰 값을 찾아 출력하시오. (힌트: max() 없이 직접 비교해보기)

```python
numbers = [3, 7, 2, 9, 5]
maximum = numbers[0]

for n in numbers:
    if maximum < n:
        maximum = n

print(maximum)
```

# 문제 3. 짝수만 저장하기
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

짝수만 새로운 리스트에 담아서 출력하시오.

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
evens = []

for n in numbers:
    if n % 2 == 0:
        evens.append(n)

print(evens)
```

# 문제 4. 이름 존재 여부 확인
target = "민수"

names = ["철수", "영희", "민수", "지영"]

변수 target 에 저장된 이름이 리스트 안에 있으면  "존재합니다" 아니면 "없습니다" 출력

```python
target = "민수"
names = ["철수", "영희", "민수", "지영"]

# in 연산자로 쉽게도 가능!
# print(target in names)

result = '없습니다'

for name in names:
    if name == target:
        result = '존재합니다'

print(result)
```

# 문제 6. 재고 확인 시스템

inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

각 과일의 재고가 0이면 → "품절" 

1 이상이면 → "판매 가능" 

```python
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

for i in inventory:
    if inventory[i] > 0:
        print(f'{i}: 판매 가능')
    else:
        print(f'{i}: 품절')
```

# 문제 7. 리스트 안 딕셔너리 평균 구하기

students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

전체 평균 점수 구하기

평균 이상인 학생 이름 출력

```python
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

total = 0
result = []

for student in students:
    total += student["score"]

average = total / len(students)
print(f'평균 점수 : {average}')

for student in students:
    if student["score"] >= average:
        result.append(student["name"])

print(result)
```

# 문제 8. 등급 매기기

점수 리스트: scores = [95, 82, 67, 74, 88]

각 점수에 대해

90 이상 → A

80 이상 → B

70 이상 → C

그 외 → D

등급 리스트를 새로 만들어 출력하시오.

```python
scores = [95, 82, 67, 74, 88]
grades = []

for score in scores:
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    else:
        grade = 'D'
    grades.append(grade)
    
print(grades)
```

# 문제 9. 장바구니 총 금액 계산

cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]

총 결제 금액 계산 100만원 이상이면 "고액 구매자" 출력

```python
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]

sum = 0

for i in cart:
    sum += i["price"] * i["qty"]

print(sum)

if sum >= 1e6:  # 1000000
    print('고액 구매자')
```

# 문제 10. 비밀번호 검사기

password = "abc123"

조건: 길이가 8자 이상이고, 숫자가 포함되어 있으면 "사용 가능", 아니면 "사용 불가"

(힌트: len(), in  사용. for 문으로 한글자씩 체크) (편집됨)

```python
password = "abc123"
possible = False

# 8글자 이상
if len(password) >= 8:
    for p in password:
        # if p.isdigit():
        # if ord(p) >= 48 and ord(p) <= 57:     # ascii 코드 값으로 확인
        if p in '0123456789':
            possible = True

print(possible)
```
