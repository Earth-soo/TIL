## 변수명(식별자)
1. 대소문자 다 됨, 대소문자 구분 됨
1. 한글 됨(근데 안 씀)
1. 미리 예약된 특수 단어들 안 됨 EX. import
1. 미리 점유된 이름은 쓸 수 있음(근데 쓰면 안 됨) EX. print

### Convention
1. 일반 변수명은 모두 소문자, `_` 로 구분
1. 일반 함수명도 "
1. 상수는 모두 대문자
1. 클래스는 첫글자만 대문자

```python
# 변수는 박스
# 변수 a에 10이라는 값을 저장한다.

a = 10
```

```python
# i : 5, j : 3, s : '파이썬'
i = 5;
j = 3;
s = '파이썬'

print(i, j, s);
```

```python
# 문제
# x = 10, y =20 일 때, 값을 서로 교체하는 코드를 작성해 보자.

x, y = 10, 20;
# x, y = y, x; # 이렇게도 가능

# 기본
# tmp = x;
# x = y;
# y = x;

# 변수 생성 없이, x와 y값 바꾸기
x = x + y;  # x: 30, y: 10
y = x -y;   # x: 30, y: 20
x = x -y;   # x: 10, y: 20

print(x, y);
```

## 자료형 (Data Type)

### Boolean (True/False)

### Numeric (숫자)
- int(정수)
- float(실수)
- complex(복소수)

### String (문자)
문자열 표현법(쌍 & 홑 따옴표 모두 가능)
- `'a'`
- `"a"`
- `'''a'''` (docstring -> 문서 설명 글귀 or 프롬프트용)
- `"""a"""`

### None (없음)


```python
# Boolean
type(True), type(False)

# bool() 함수 : 특정 데이터가 True, False 인지 검증
bool(0), bool(-100), bool(10), bool('hello'), bool(''), bool([1, 2, 3]), bool([])
```

```python
# int
i = 3
type(i)     # int

i = 2 ** 100    # 파이썬은 메모리 오버플로우 걱정 없음
print(i)    

# float
f = 3.5   
type(f)     # float

type(3), type(3.1), type(3 * 3.1), type(3 + 3.0)    # int, float, float, float

a = 3.5 - 3.12    # 순환소수의 오차 때문에 0.38 이 나오지 않음

import math
math.isclose(0.38, a)   # 소수 오차범위 안에서 같은 수인지 판단
```

```python
# string
s = 'hello'
h = 'python'
print(s + h)

ss = '안녕하세요\n반갑습니다.'
print(ss)
# '안녕하세요
# 반갑습니다.'      # `\n`로 해서 한 줄 안에 여러줄로 표시되도록 or `'''`로 여러줄로 표시

print()     # '\n' 만 출력

ds = '''안녕하세요
반갑습니다
내일은 눈이 올 거 같아요'''
print(ds)
```

```python
name = '헝'
age = 24

# greeting = name + '님 환영합니다.' + name + '님의 나이는 ' + age + '입니다.'  # 너무 복잡함

# f-string : 글자 안에 변수 녹여 쓰기(f'hi', f'''hi''' 둘 다 가능)
greeting = f'''{name}님 환영합니다. {name}님의 나이는 {age}살 입니다.'''
print(greeting)

# %포맷팅(예전 style)
print("당신의 이름은 %s이고 나이는 %d살입니다." % (name, age))
```

```python
# String 안에서 따옴표 사용하기(`\` 사용!)
title = 'It\'s my bag.'     # "It's my bag" 도 가능
story = '그는 말했다. \'안녕?\' 이라고'
print(title, story)
```
