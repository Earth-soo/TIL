# 연산자(operator)

## 산술연산자

|연산자|내용|
|---|---|
|`+`|더하기|
|`-`|빼기|
|`*`|곱하기|
|`/`|나누기|
|`//`|몫|
|`%`|나머지|
|`**`|거듭제곱|

```python
# 2의 1000 제곱
2 ** 1000

# 정수 - 정수 연산은 대부분 정수지만, '/'는 실수가 나온다
x = 3 / 2
y = 4 / 2

x, y

a = 100
-a, +a  # 음수, 양수 표시
```

## 비교 연산자

|연산자|내용|
|---|---|
|`>`, `<`|초과, 미만|
|`>=`, `<=`|이상, 이하|
|`==`|값이 같음|
|`!=`|다르다|
|`in`|포함여부|
|`is`|동일하다|

```python
5 > 3   # T
4 > 2   # F
5 >= 5   # T
5 <= 3   # F

5 == 3  # F
4 != 2  # T

3.5 - 3.12 == 0.38  # F(순환소수의 특성 때문)

'apple' == 'apple'  # F 

# x in y : y 안에 x 가 있는지
'i' in 'apple'  # F
3 in [1, 2, 3, 4]   # T
```

## 논리 연산자

|연산자|내용|
|---|---|
|`and`|`a and b` -> a, b 모두 True일 경우에만 True|
|`or`|`a or b` -> a, b 모두 False일 경우에만 False|
|`not`|`not True` -> 값을 뒤집는다|

```python
# and
True and True
True and False
False and True
False and False

# or
True or True
True or False
False or True
False or False

# not
not True
not False
```

### Short-Circuit Evaluation (단축 평가)
- 앞에서 값이 확실 시 되면, 그 뒤는 확인하지 않음
- 조건 판별 시 뒷부분을 확인하지 않아도 됨
- 결정이 된 부분, 즉 마지막으로 본 값으로 표시 됨

```python
x = 1 and [] and 3 and 0 and 4   # 값이 없을 때의 값으로 표시([]일 때 False가 되면서 []으로 표시)
y = 0 or None or '' or [] or [None] or 10 or 'asdf'     # 값이 있을 때의 값으로 표시([None]일 때 True가 되면서 [None]으로 표시)

print(x, y)

test = '0211'
test or '기본값'    # test 값이 있으면 test 값 사용, 없으면 '기본값' 사용
```

## 복합 연산자

|연산자|내용|
|---|---|
|`a += b`|`a = a + b`|
|`a -= b`|`a = a - b`|
|`a*= b`|`a = a * b`|
|`a /= b`|`a = a / b`|
|`a //= b`|`a = a // b`|
|`a %= b`|`a = a % b`|
|`a **= b`|`a = a ** b`|

```python
x = 1
x += 1  # x = x + 1 과 같은 의미
print(x)

s = 'py'
s += 'thon'
print(s)
```

```python
# bmi 계산기
# bmi = 몸무게(kg) / 키(m) 제곱 2
# weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기

weight = 45
height = 1.56
bmi = weight / (height ** 2)

print(bmi)
```

```python
# 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
# 원 넓이 = pi x r^2
# 원 둘레 = 2 x pi x r
from math import pi

r = 3

w = round(pi * (r ** 2), 2)
l = round(2 * pi * r, 2)

print(f"지름이 {r}인 원의 넓이는 {w}, 원의 둘레는 {l} 입니다.")
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환

seconds = 3675  # 1시간 1분 15초

h = seconds // 60 // 60
m = seconds // 60 % 60
s = seconds % 60

# h, seconds = divmod(seconds, 3600)      # (1, 75)
# m, seconds = divmod(seconds, 60)        # (1, 15)
# s = seconds                             # 15

# h = seconds // 3600
# seconds %= 3600

# m = seconds // 60
# seconds %= 60

# s = seconds

print(f"{seconds}는 {h}시간 {m}분 {s}초 입니다.")
```

```python
# 세자리 수 분리하기
# 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기

n = 472

i = n // 10 ** 2
ii = (n % 100) // 10
iii = n % 10

print(f"{n}의 백의 자리는 {i}, 십의 자리는 {ii}, 일의 자리는 {iii} 입니다.")
```

```python
a = 0b1010   # 2진수 → 10 (0b로 표현)
b = 0o12     # 8진수 → 10 (0o로 표현)
c = 10       # 10진수
d = 0xA      # 16진수 → 10 (0x로 표현)
e = 2e4      # 20000, 부동소수점 지수 표현에서 사용 (e는 10을 의미)

print(a, b, c, d, e)

itob = bin(10)  # '0b1010'
itoo = oct(10)  # '0o12'
itox = hex(10)  # '0xa'

print(itob, itoo, itox)
```
