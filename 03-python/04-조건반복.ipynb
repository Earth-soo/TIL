```python
# 조건 - if
n = 10

if n > 0:
    print('양수 입니다.')
    print('11111')

# 블럭 밖은 전혀 다른 문
print('2222222')

# 조건 - if-else
if n > 0:
    print('0보다 크다.')
else:
    print('0보다 작거나 같다.')

# 조건 - if-elif-else
if n > 0:
    print('양수')
elif n == 0:
    print('0이다.')
else:
    print('음수')
```

```python
# 불합: 0-399
# 1: 400-599
# 2: 600-749
# 3: 750-899
# 4: 900-1000
score = 980

if score >= 900:
    print('level 4')
elif score >= 750:
    print('level 3')
elif score >= 600:
    print('level 2')
elif score >= 400:
    print('leve 1')
else:
    print('ㅠㅠ')
```

```python
# 우수 학생 판별
status = {
    'score': 80,
    'absent': 3
}

# 1. 정수가 70을 넘는 학생이면 우수학생
# 2. 우수학생 중에 결석일이 5일미만이면 최우수 학생
# 3. (우수 아닌) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반 학생"

if status['score'] >= 70:
    if status['absent'] < 5:
        print('최우수 학생')
    else:
        print('우수 학생')
elif status['absent'] < 5:
    student = '성실학생'
else:
    student = '일반 학생'
```

## 반복
- while
- for

```python
#
n = 0
while n < 5:
    print('출력', n)
    n += 1

print('끝나고', n)

#
numbers = [1, 2, 3, 4, 5]
total = 0
i = 0

while i < len(numbers):
    total += numbers[i]
    i += 1

print(total)
```

## for .. in 구문

```python
numbers = [1, 2, 3, 4, 5]
total = 0

for num in numbers:
    total += num

print(total)

# for & range : firecrawl 용
for n in range(100):
    print(n)
```

```python
# 리스트는 idx 수정 가능, 문자열은 idx 수정 안 됨
# string도 list처럼 사용 가능
for c in 'python':
    print(f'!{c}')
```
