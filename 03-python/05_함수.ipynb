````python
# name: plus_one / in: 숫자 / out: 숫자 + 1

# 함수 정의(define)
def plus_one(num):  # 정의할 땐 변수(parameter)
    return num + 1

# 함수 호출(call)
plus_one(99)    # 실행할 때는 인자(argument)
````

````python
# n번 인사하는 함수
# name: hello_ntimes

def hello_ntimes(n):
    for _ in range(n):      # for "~" 의 변수를 안 쓰면 `_` 로 표시
        print('안녕하세요!')


hello_ntimes(5) # 5번 hello를 출력
````

````python
# 3제곱 해주는 함수
# name: cube, in: 숫자 1개, out: 숫자의 세제곱
def cube(n):
    return n ** 3

# 함수 호출
cube(3)
````

````python
# 원기둥의 부피를 구하는 함수
# name: cylinder, in: 반지름/높이, out: 부피
from math import pi

def cylinder(r, h):
    volume = pi * (r ** 2) * h
    return volume

print(cylinder(2, 5))  # 반지름: 2, 높이: 5
print(round(cylinder(5, 2), 2))  # 반올림
````

````python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in: 숫자들로 만들어진 리스트, out: 평균
def get_avg(numbers):
    total = 0   
    for n in numbers:
        total += n

    return total / len(numbers)

    # avg = sum(numbers) / len(numbers)
    # return avg

get_avg([1, 2, 3, 4, 5, 23, 45, 88]) 
````

````python
# 비밀번호 체크해 주는 함수 (숫자 포함, 8글자 이상)
# name: check_pw, in: 비밀번호 문자열, out: True/False
def check_pw(pw):    
    if len(pw) >= 8:
        for p in pw:
            if p.isdigit(): # if p in '0123456789':
                return True
    return False

print(check_pw('abcd1234'))    # True
print(check_pw('abcdefg'))    # False
````

````python
# BMI 기준으로, 비만여부 확인해주는 함수 (BMI = 체중(kg) / 키(m)^2)
# 남성 27 이상일 경우 비만
# 여성 25 이상일 경우 비만

def check_bmi(weight, height, gender):
    bmi = weight / (height/100) ** 2
    result = '정상'

    if gender == 'M':
        if bmi >= 27:
            result = '비만'
    elif gender == 'F':
        if bmi >= 25:
            result = '비만'
    
    return result


# 몸무게, 키, M -> male
print(check_bmi(100, 175, 'M'))  # 비만
# 몸무게, 키, F -> female
print(check_bmi(50, 160, 'F'))  # 정상
````
