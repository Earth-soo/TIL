## List(리스트) : 값들이 여러개 있으면 변수명+**`s` 필수**

```python
numbers = [1, 2, 3, 4, 5]

numbers[0]  # 1
numbers[2]  # 3

# -1 ~ 리스트 길이
numbers[-1] # 5

# # 불가능 (인덱스 범위 초과)
# numbers[-6]
# numbers[5]

# 슬라이싱 [x:y] -> x부터 y전까지
numbers[0:3]    # 1, 2, 3
numbers[1:1]    # x
numbers[0:5]    # 1, 2, 3, 4, 5 (전부)
numbers[:3]     # 1, 2, 3 (비어있음 : 끝부터)
numbers[2:]     # 3, 4, 5 (비어있음 : 끝까지)
numbers[:]      # 1, 2, 3, 4, 5 (전부)

numbers == numbers[:]   # 값은 같음
numbers is numbers[:]   # 같은 존재는 아님

# 리스트 수정
numbers[2] = 300    # 특정 index의 데이터 바꾸기
numbers[-1] = 500

# 리스트에 데이터 추가
numbers.append(6)   # 마지막에 6 추가

# 리스트의 데이터 삭제하기
numbers.pop(-1)     # 6 삭제
numbers.pop()       # 그냥은 마지막?
```

## Range(범위)

```python
r1 = range(1, 10)   # 1 ~ 9 (처음 이상 ~ 마지막 미만)
# 눈으로 보려면, 리스트로 바꿔야함
print(list(r1))

# 시작 숫자 X -> 0부터 시작
r2 = range(5)
print(list(r2))

# step 개념
r3 = range(0, 10, 2)    # 0, 2, 4, 6, 8 (처음 이상 ~ 마지막 미만까지 step 씩 넘기)
print(list(r3))

# 음수 step(뒤로 가기)
r4 = range(5, 0, -1)     # 5, 4, 3, 2, 1 (시작, 끝, 방향 + 스텝)
print(list(r4))
```

## Dict (딕셔너리)
| key - Value
- JSON : JavaScript Object Notation 의 줄임말
    - [JS언어의 Object 자료형 식 표기법]
- Python 에는 Dict 자료형    

```python
# dict 생성
me = {
    'name': 'neo',
    'age': 20,
    'is_hungry': True,
    'height': 123.4,
    'skills': ['n8n', 'sql', 'python']
}

# 접근 -> 대괄호 안에 key를 쓴다
me['name'], me['age']

# 'python' 에 접근(key값으로 접근)
me['skills'][2]     # js에서는 `.`, [''] 둘 다 가능 -> $json.data.list[0] 랑 $json["data"]["list"][0] 가능

# 변경
me['name'] = '김새싹'
me['is_hungry'] = False

type(me['skills'])      # list (# me['skills']는 리스트, 리스트에 아이템 추가는 append())
me['skills'].append('git')

# K-V 추가
me['money'] = 10000
print(me)

# K-V 삭제
me.pop('height')

print(me)
```

## Set(집합)
- 중복되는 요소가 없다
- 순서(index)가 없다.

```python
s1 = {1, 1, 2, 3, 4, 2}
s2 = {3, 4, 5, 5}

# 합집합
print(s1 | s2)
# 교집합
print(s1 & s2)
# 차집합
print(s1 - s2)

# 바꿔서 편하게 쓰기 가능
list(s1)

# 중복이 많은 리스트에서 중복을 제거하는 용도로 쓰임
ll = [1,2,3,4,52,1,12,3,31,1,1,1,1,1,2,2,2,3,4,4,4,5,2,1,12]
ss = set(ll)    # set으로 변환 시 순서가 없어짐
ll = list(ss)
ll.sort()   # 오름차순 정렬
print(ll)
```

## Tuple (튜플)
- 리스트 친척
- 순서가 있음(index 접근 가능)
- 리스트와 차이 : 수정 불가능
- 직접 쓰기 보단 python 내부에서 많이 쓰임

```python
t = (10, 20, 30, 40, 50)

print(t[0:3], t[-1], t[:])

# 수정 불가능
# t[0] = 100
# t.append()
# t.pop()
```
