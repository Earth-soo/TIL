### OOP(객체지향언어) -> 실존하는 값이 주인공
- Object : 객체 : 값(`1`, `'abc'`, `[1, 2, 3]`)
- Oriented : 지향 : 핵심, 주인공

- Class(붕어빵 틀/개념) - Instance(붕어빵/실체)
- 값이라고 부를 수 있는 것은 Instance

```python
# Person
# 키, 몸무게, 나이, 이름
# 인사, 공부

# Python에서 클래스 이름은 PascalCase로 짓는다  my_good_friend / MyFoodFriend
class Person:
    # 1. 특성(attribute)
    height = 165
    weight = 55
    age = 20
    iq = 100
    name = '인간'
    
    # 2. 행동(method)
    def greeting(self):
        return f'안녕하세요. {self.name} 입니다. {self.age} 살 입니다.'
    
    def study(self):
        self.iq += 5
```

```python
# Person 개념적 대상 -> 실제 생성
p1 = Person()
p2 = Person()

# 공장 초기화 된 정보
print(p1.name)
print(p2.name)

# 특성 재설정
p1.age = 30 
p1.name = '김로봇'
p1.weight = 80
p1.height = 230

# 행동
print(p1.greeting())

p2.age = 23
p2.name = '박사람'
p2.weight = 45
p2.height = 155
p2.greeting()
```

```python
r1 = range(100)     # range 클래스의 인스턴스를 만듦
l1 = list(r1)       # list 클래스의 인스턴스를 만듦

'abc'   # string 클래스의 인스턴스 -> string 으로 부르고 있음
a = 3 / 1   # a의 인스턴스 : 3.0, a의 클래스 : float
type(a)     # type의 역할 : 인스턴스의 클래스를 나타내는 것
```

```python
# 객체 지향 세상 -> 객체들끼리 상호작용(메서드) 하는 세상

(2).__add__(3)
2 + 3   # 위의 객체들의 상호작용을 보다 쉽게 만든 것
```
