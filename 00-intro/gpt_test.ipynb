```python
# Appliation -> 프로그램 1. gpt / 2. 지금 우리가 만드려고 하는 것
# OPENAI API 사용하기 -> GPT한테 질문하고, 답을 받겠다.
# 1. OPENAI 와 통신할 SW 설치
%pip install openai
```

```python
prompt = """
2025년 서울과 경기도 지역에 있는 기업의 풀스택, AI, 프론트, 백 4분야의 채용 공고량과 인원 조사 해줘.
"""

from openai import OpenAI
client = OpenAI(api_key='')
# openAI는 사용할 때마다 금액이 청구, api_key로 결제 키 값이 있어야 처리됨.

response = client.responses.create(
    model = 'gpt-5.2', # 모델 변경 가능
    input = prompt
)

print(response.output_text)
```
