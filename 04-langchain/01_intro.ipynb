```python
from langchain.agents import create_agent
from dotenv import load_dotenv

load_dotenv()

    # 매개변수 : 매개변수의 타입 -> return 타입
def get_weather(city: str) -> str:
    """Get weather for a given city."""     # <- docstring : agent가 어떤 함수인지 알 수 있도록 만든 설명서("""로 구성 필수)
    return f"It's always sunny in {city}!"


def calculator(num1:float, num2:float, opr:str) -> float:
    """두 수의 사칙연산을 진행"""
    if opr == "+":
        result = num1 + num2
    elif opr == "-":
        result = num1- num2
    return result

agent = create_agent(
    model="openai:gpt-4.1-mini",
    tools=[get_weather, calculator],    # tool-calling : agent에게 tool을 쥐어주고 필요 시 tool call 요청
    system_prompt="You are a helpful assistant",
)

# Run the agent
agent.invoke(
    {"messages": [{"role": "user", "content": "What is the weather in San Francisco? and calculate 3 + 3."}]}
)
```
